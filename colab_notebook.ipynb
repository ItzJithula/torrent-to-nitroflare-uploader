## 3. Configure your upload backends

You can upload files to **Nitroflare** and/or **Gofile.io**.

- **Nitroflare**: Get your user hash from https://nitroflare.net/user-hash
- **Gofile.io**: Get your API token from https://gofile.io/myprofile (leave empty for guest uploads)

```python
NITROFLARE_KEY = 'YOUR_NITROFLARE_USER_HASH_HERE'  # <-- paste your hash (or keep as-is for Gofile only)
GOFILE_TOKEN = 'YOUR_GOFILE_API_TOKEN_HERE'        # <-- paste your Gofile token (or keep empty for guest uploads)

import yaml
with open('config.example.yaml') as f:
    cfg = yaml.safe_load(f)
cfg['nitroflare']['api_key'] = NITROFLARE_KEY
cfg['gofile']['api_token'] = GOFILE_TOKEN
with open('config.yaml', 'w') as f:
    yaml.dump(cfg, f)
print('config.yaml written')print('  Nitroflare:', 'configured' if NITROFLARE_KEY != 'YOUR_NITROFLARE_USER_HASH_HERE' else 'not set')print('  Gofile.io:', 'configured' if GOFILE_TOKEN != 'YOUR_GOFILE_API_TOKEN_HERE' else 'guest mode (no token)')
```
